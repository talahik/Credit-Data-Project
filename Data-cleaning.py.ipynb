```python
#Lire le fichier "german credit data.csv" dans un Dataframe applé df sachant que la premiere ligne contient les indices
df=pd.read_csv("german credit data.csv", sep=";", index_col=0)
```

```python
#Le montant maximum du crédit suivant les différent motifs 
df.groupby(["purpose"])["credit_amount"].max()
```

```python
#Determiner le motif revenant le plus souvent
df["purpose"].value_counts(normalize= True)
```

```python
#Affichage des informations concernant l'individue le plus agé
print("Individue le plus agé: ")
print(df.sort_values(by="age", ascending = False).iloc[0])      
#Affichage  des informations concernant l'individue le plus jeune
print("Individue le plus jeune: ")
print(df.sort_values(by="age", ascending= True).iloc[0])
```

```python
#Transformation de la variable "duration" en variable numérique en supprimant la sous chaine "month"
df['duration'] = df['duration'].apply(lambda x: int(x[:-6]) if isinstance(x, str) else x)
#Verification 
df.head()
```

```python
#Creation d'une nouvelle variable nommé "duration-categ"
def filtre(x):
    if x >= 0 and x <= 10 :
        return "court-terme"
    if x > 10 and x <= 30 :
        return "moyen-terme"
    if x > 30 :
        return "long-terme"
df["duration_categ"]= df["duration"].apply(filtre)
```

```python
df.head()
```

```python
#Determiner le type de prêt le plus commun chez les locataires
pd.crosstab(df["housing"], df["duration_categ"])
```

## Préparation des données 

```python
#Remplacement des modalités "little","moderate","quite rich" et "rich" par 0,1,2,3 dans la colonne "saving_accounts"
df["saving_accounts"]=df["saving_accounts"].replace(["little", "moderate", "quite rich", "rich"], [0, 1, 2, 3])
```

```python
#Verification
df.head()
```

```python
#Remplacement des modalités "little","moderate","quite rich" et "rich" par 0,1,2,3 dans la colonne "cheking_account"
df["checking_account"]= df["checking_account"].replace(["little","moderate","quite rich", "rich"], [1,2,3,4])
```

```python
#Verification
df.head()
```

```python
#Verifiaction de présence des doublons dans les données
df.duplicated().sum()
```

```python
#Affichage de nombre des valeurs manquantes pour chaque colonne de df
df.isna().sum()
```

```python
# Remplacer les valeurs manquantes de la colonne "housing" par le mode
df["housing"]= df["housing"].fillna(df["housing"].mode()[0])
```

```python
#Verification
df["housing"].isna().sum()
```

```python
#Remplacer les valeurs manquantes de la colonne "age" par la moyenne
df["age"]= df["age"].fillna(df["age"].mean())
```

```python
#Verification
df["age"].isna().sum()
```

## Apprentissage des données 

```python
# Lire le fichier "Je predictions german.csv"
pred_german = pd.read_csv("predictions german.csv", sep=";", index_col =0)
```

```python
#Affichage des premières lignes de pred_german
pred_german.head()
```

```python
#Fusion du df et pred_german
df_pred = df.join(pred_german)
```

## Evaluation du modèle de régression

```python
#Création d'une nouvelle colonne "error" renseignant la différence entre les variables "duration" et "prédictions"
df_pred["error"]= df_pred["duration"] - df_pred["predictions"]
```
